```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# Definir los puntos
x = np.array([1, 2, 3])
y = np.array([2, 3, 5])

# Crear el Spline cúbico
cs = CubicSpline(x, y)

# Evaluar el Spline en el punto x = 2.5
x_eval = 2.5
y_eval = cs(x_eval)
print(f"Valor interpolado de y para x = 2.5: {y_eval}")

# Gráfico comparativo
x_plot = np.linspace(1, 3, 100)
y_plot = cs(x_plot)

plt.plot(x, y, 'o', label='Puntos dados')
plt.plot(x_plot, y_plot, label='Spline cúbico')
plt.scatter(x_eval, y_eval, color='red', label=f'Interpolación en x=2.5, y={y_eval}')
plt.legend()
plt.title('Interpolación por Spline cúbico')
plt.show()
```
